```python
# 13. classify movies based on rating (excellent, good and average)

def rating (rating):
    if rating>=7.0:
        return 'Excellent'
    elif rating>=6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
df['Rating_cat']= df['Rating'].apply(rating)
```
