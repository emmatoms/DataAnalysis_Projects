## Spotify Daily Chart - Nigeria - 2024/02/22

### Web Scrapping form Kworb.net

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://kworb.net/spotify/country/ng_daily.html"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, "html")
```

```python
print(soup.prettify())
```

```python
table = soup.find("table", class_="sortable")
print(table)
```

```python
table.find_all("th")
```

```python
chart_table_title = table.find_all("th")
```

```python
chart_table_title = [title.text for title in chart_table_title]
print(chart_table_title)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = chart_table_title)
```

```python
df
```

```python
column_data = table.find_all("tr")
print(column_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all("td")
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
    
    length = len(df)
    df.loc[length] = individual_row_data
```
