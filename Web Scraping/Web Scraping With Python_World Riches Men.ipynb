## Web scraping from music.apple.com

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://beinsure.com/world-richest-people/"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, "html")
```

```python
print(soup.prettify())
```

```python
table = soup.find("table")
table
```

```python
table_title = table.find('tr')
table_title
```

```python
richmen_table_title = [title.text for title in table_title]
richmen_table_title
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = richmen_table_title)
df
```

```python
column_data = table.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
    
    length =len(df)
    df.loc[length] = individual_row_data
```

```python
df.set_index('№')
```
